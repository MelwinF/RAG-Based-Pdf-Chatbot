Streamlit-style logic in notebook

```python
from transformers import pipeline
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
```

```python
file_name = 'sample.pdf'  # replace with your file
loader = PyPDFLoader(file_name)
documents = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
docs = splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2')
db = FAISS.from_documents(docs, embeddings)
```

```python
generator = pipeline('text-generation', model='tiiuae/falcon-rw-1b')
```

```python
def ask_question(query):
    results = db.similarity_search(query, k=5)
    context = ' '.join([doc.page_content for doc in results])
    prompt = f"""Answer using context:\n{context}\nQuestion:{query}\nAnswer:"""
    response = generator(prompt, max_new_tokens=100)
    print(response[0]['generated_text'])
```
